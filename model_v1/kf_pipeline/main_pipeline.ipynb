```python

#import kfp
from kfp.v2 import dsl

from kfp.v2.dsl import (Artifact,
                        Dataset,
                        Input,
                        Model,
                        Output,
                        Metrics)

from kfp.v2 import compiler


from pipeline.basic_preproc import basic_prepoc
from pipeline.token_stopw_preproc import token_stopw_preproc
from pipeline.emb import emb_model
from pipeline.cluster import cluster_model
```

```python
root_path = "gs://asl-project-globo-de18f040"

@dsl.pipeline(
    pipeline_root=root_path ,
    name="pipeline-test-1")


def pipeline(root_path:str=root_path):
    
    basic_clean_dataset = basic_prepoc(root_path=root_path)
    
    dataset_token_stopw = token_stopw_preproc(in_dataset=basic_clean_dataset.outputs["dataset"])
    
    emb_txt = emb_model(in_dataset=dataset_token_stopw.outputs["out_dataset"])
    
    cluster_txt = cluster_model(in_dataset=emb_txt.outputs["out_dataset"])
    
```

```python
compiler.Compiler().compile(pipeline_func=pipeline,
        package_path='pipe_asl_project.json')
```
